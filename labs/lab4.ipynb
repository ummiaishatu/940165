```python
import numpy as ny
import matplotlib.pyplot as mat 
```

```python
data = ny.load('wineData.npy')
labels = ny.load('wineLabels.npy')
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_data, test_data, train_labels, test_labels = train_test_split(data, labels, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
traindata = scaler.fit_transform(train_data)
testdata = scaler.transform(test_data)

print(f'Minimums of the features are: {traindata.min(0)}')
print(f'Maximums of the features are: {traindata.max(0)}')

mat.figure()
mat.scatter(traindata[:,0], traindata[:,1], c=train_labels)
mat.xlabel('Feature Dimension 0')
mat.ylabel('Feature Dimension 1')
mat.title('Standardised Data')
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC()
svc_train = svc.fit(train_data, train_labels)
svc.score(test_data, test_labels)
predictedlabel = svc.predict(test_data)
```

```python
mat.figure(figsize=[10,9])
mat.subplot(121)
mat.scatter(test_data[:,0], test_data[:,1], c=test_labels)
mat.ylabel('Feature Dimension 1')
mat.xlabel('Feature Dimension 0')
mat.title('Ground Truth')
mat.subplot(122)
mat.title('Predicted Labels')
mat.xlabel('Feature Dimension 0')
mat.scatter(test_data[:,0], test_data[:,1], c=predictedlabel)
mat.show(block=False)
```
