## PRINCIPLE COMPONENT ANALYSIS (PCA) TASK 3.2 

```python
data = ny.load('wineData.npy')
labels = ny.load('wineLabels.npy')
print(data)
print(labels)
mat.scatter(data[:, 0], data[:, 1], c=labels)
mat.xlabel('data 1')
mat.ylabel('data 2')
mat.title('Wine Data sample')
mat.show(block=False)
```

```python
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
```

```python
n = data.shape[0]
split = int(n * 0.8) 

train_data = data[:split,:]
train_labels = labels[:split]

test_data = data[split:,:]
test_labels = labels[split:]

mat.scatter(train_data[:,0], train_data[:,1], marker = 'o', c=train_labels)
mat.scatter(test_data[:,0], test_data[:,1], marker = 'x', c=test_labels)
mat.xlabel('Feature 0')
mat.ylabel('Feature 1')
mat.title('Wine Data first scatter plot')
mat.show(block=False)
```

```python
pca = PCA(n_components=2)
pca.fit(train_data)

dimensiontraindata = pca.transform(train_data)

dimensiontestdata = pca.transform(test_data)

mat.scatter(dimensiontraindata[:,0], dimensiontraindata[:,1], marker = 'o', c=train_labels)
mat.scatter(dimensiontestdata[:,0], dimensiontestdata[:,1], marker = 'x', c= test_labels)
mat.xlabel('PC 0')
mat.ylabel('PC 1')
mat.title('Wine Data for Principle Component Analysis')
mat.show(block=False)
```

## Task 3.4 done by hand

```python
meantrain = ny.mean(train_data, axis = 0)
#69.89510075839652
center1 = train_data - meantrain
print(center1)
u1, s1, vh1 = ny.linalg.svd(center1)
```

```python
projection_matrix = vh1
projected_train_data = center1 @ ny.transpose(projection_matrix)
```

```python
center2 = test_data - meantrain
projected_test_data = center2 @ ny.transpose(projection_matrix)
```

```python
mat.scatter(projected_train_data[:,0], projected_train_data[:,1], marker = 'o', c=train_labels)
mat.scatter(projected_test_data[:,0], projected_test_data[:,1], marker = 'x', c= test_labels)
mat.xlabel('Projection 0')
mat.ylabel('Projection 1')
mat.title('Principal Component Analysis by hand')
mat.show(block=False)
```
